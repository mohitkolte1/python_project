# PIVOT

```python
df = pd.DataFrame({'id':[101,102,103,104],
                   'name':['a','b','c','d'],
                   'date':['01-04-2024','04-05-2023','03-02-2024', '05-02-2024'],
                  'sal':[12313132,1232132,123123123,123123123]})
```

```python
# index as date mostly in real time bcz easy to filter
```

```python
## data --> id name?
df.pivot(index='date', columns='sal')
```

```python
df = pd.DataFrame({'dept':['IT','Admin','Finance','HR'],
                   'date':['01-04-2024','04-05-2023','03-02-2024', '05-02-2024'],
                  'sal':[12313132,1232132,123123123,123123123]})
```

```python
df
```

```python
df = pd.DataFrame({'city':['Pune','Mumbai','Pune','Mumbai'],
                   'date':['01-04-2024','04-05-2023','03-02-2024', '05-02-2024'],
                  'temp':[12,15,16,12]})
```

```python
df.pivot(index='date', columns="city")
```
